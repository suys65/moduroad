```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
import pandas as pd
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import random
import traceback
# Function to simulate random delay
def random_delay(min_seconds, max_seconds):
    """
    Pauses the script for a random interval between min_seconds and max_seconds.
    """
    time.sleep(random.uniform(min_seconds, max_seconds))


# ChromeDriver의 경로를 지정합니다. 여러분의 환경에 맞도록 수정하세요.
webdriver_path = 'c:/my_code/croller/chromedriver.exe'

# Selenium WebDriver 초기화
service = Service(webdriver_path)
driver = webdriver.Chrome(service=service)

# 웹페이지 로드
driver.get('https://icloud.incheon.go.kr/arcgis/home/item.html?id=651b6ae622374513b57b5cdc9cdc01c7&view=list&showFilters=true&sortOrder=desc&sortField=defaultFSOrder#data')


# 웹페이지 로딩을 기다림
time.sleep(5)

# <select> 요소 찾기 (수정된 방식)
select_element = driver.find_element(By.CSS_SELECTOR, "#esri_Evented_0 > div.padding-leader-half.padding-trailer-half.clearfix.panel-white > div > div.column-16.tablet-column-6.phone-trailer-half.flex > div > label > select")

# Select 객체 생성
select_object = Select(select_element)

# value가 '특정값'인 <option> 선택하기
select_object.select_by_value('7')

data = []
start_index = 1  # Assuming you're starting from item 800

# Focus on the scrollable element
#scrollable_element = WebDriverWait(driver, 10).until(
 #   EC.presence_of_element_located((By.CSS_SELECTOR, "#dgrid_1-row-1 > table > tr > td.dgrid-cell.dgrid-cell-padding.dgrid-column-9.field-endlongitude"))
#)
#scrollable_element.click()  # Ensure the element is focused to receive keyboard inputs

try:
    for i in range(start_index, 84): 
        if i % 10 == 0:
            print(i) # Adjust the range as needed
        # Send the PAGE_DOWN key to scroll
            css_selector11 = f'#dgrid_1-row-{i} > table > tr > td.dgrid-cell.dgrid-cell-padding.dgrid-column-0.field-ctprvnnm'
            #css_selector11 = f'#dgrid_1-row-{i} > table > tr > td.dgrid-cell.dgrid-cell-padding.dgrid-column-0.field-prkplceno'
            scrollable_element = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, css_selector11))
            )
            scrollable_element.click() 
            
            scrollable_element.send_keys(Keys.PAGE_DOWN)
            random_delay(0.5, 1)# Introduce a random delay between key presses

        css_selector = f'#dgrid_1-row-{i} > table'
            
        try:
            wait = WebDriverWait(driver, 10) # 최대 10초간 대기
            
            table = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, css_selector)))
            rows = table.find_elements(By.TAG_NAME, 'tr')
            
            for row in rows:
                row_data = []  # 현재 행의 데이터를 저장할 리스트
                cells = row.find_elements(By.TAG_NAME, 'td')
                for cell in cells:
                    row_data.append(cell.text)  # 현재 셀의 데이터를 리스트에 추가
                data.append(row_data)
                # 현재 행의 데이터를 전체 데이터 리스트에 추가
        except Exception as e:
            print(f"{css_selector}를 찾을 수 없습니다.", e)
            traceback.print_exc()
        

except Exception as error:
    print("An error occurred:", error)
    traceback.print_exc()

# Clean up and close the WebDriver
driver.quit()

# DataFrame 생성
df = pd.DataFrame(data, columns=['ctprvnnm','signgunm','signgucode','rdnmadr','lnmadr','latitude','longitude','institutionnm','phonenumeric','referencedate','instt_code','instt_nm','handicapcvntlyn','handicapcvntltype','un'])
#df.to_pickle('data/obstacle/stairs_steep.pkl')

# DataFrame 출력
df
```
